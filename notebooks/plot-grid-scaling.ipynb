```python
import numpy as np
import matplotlib.pyplot as plt
```

All values reported are for Eqip Sermia.

```python
area_flags = [1000**2, 800**2, 600**2, 400**2, 200**2]
a_links = [3644, 4055, 5016, 8125, 26158]
a_nodes = [1560, 1698, 2020, 3061, 9104]
a_area = [206804, 182851, 143851, 85106, 25271]
a_len = [532, 532, 511, 429, 247]

quality_flags = [5, 10, 15, 20, 25, 30]
q_links = [1983, 2264, 2671, 3167, 3888, 6244]
q_nodes = [966, 1068, 1219, 1401, 1667, 2507]
q_area = [423775, 360351, 296818, 244043, 194047, 115327]
q_len = [632, 586, 536, 494, 456, 374]
```

```python
fig, ax = plt.subplots(1, 2, figsize = (14, 4))
ax[0].plot(area_flags, a_links)
ax[0].plot(area_flags, a_nodes)
ax[1].plot(quality_flags, q_links)
ax[1].plot(quality_flags, q_nodes)
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize = (14, 8))

ax[0, 0].plot(area_flags, a_area)
ax[0, 0].set_title('Area flag vs. Mean patch area')

ax[0, 1].plot(area_flags, a_len)
ax[0, 1].set_title('Area flag vs. Mean link length')

ax[1, 0].plot(quality_flags, q_area)
ax[1, 0].set_title('Quality flag vs. Mean patch area')

ax[1, 1].plot(quality_flags, q_len)
ax[1, 1].set_title('Quality flag vs. Mean link length')

plt.show()
```
